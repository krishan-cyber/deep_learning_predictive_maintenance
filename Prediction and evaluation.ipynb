```python
from tensorflow.keras.models import load_model
import pandas as pd 
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import precision_score, recall_score, f1_score, confusion_matrix
```

```python
loaded_model = load_model('model.h5')
```

```python
x_train=pd.read_csv('data/data_processed/x_data.csv').drop(['Unnamed: 0'], axis=1)
y_train=pd.read_csv('data/data_processed/y_data.csv').drop(['Unnamed: 0'], axis=1)
```

```python
y_train
```

```python
dict_map = {'none': 0, 
            'comp1': 1, 
            'comp2': 2, 
            'comp3': 3,
            'comp4': 4, 
           }
def get_key(value):
    keys=[key for key,val in dict_map.items() if val==value ]
    return keys[0]
```

```python
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
predictions = loaded_model.predict(x_train[[4]])
```

```python
binary_predictions = (predictions > 0.5).astype(int)
print("component expected to fail:",get_key(binary_predictions[0][0]))
```

```python
y_pred=[]
```

```python
for i in range(0,5001):
    predictions = loaded_model.predict(x_train[[i]])
    y_pred.append((predictions > 0.5).astype(int))
```

```python
y_pred=np.array(y_pred)
y_train2=np.array(y_train)
y_pred=np.reshape(y_pred,(5001,1))
```

```python
precision = precision_score(y_train2[0:5001], y_pred,average='weighted')
recall = recall_score(y_train2[0:5001], y_pred,average='weighted')
f1 = f1_score(y_train2[0:5001], y_pred,average='weighted')
conf_matrix = confusion_matrix(y_train2[0:5001], y_pred)
```

```python
print(f'precision : {precision}\nRecall : {recall}\nf1 : {f1}\nconfusion matrix :\n {conf_matrix}')
```
